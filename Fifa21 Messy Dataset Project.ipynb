```python
#convert 'Height' column from object to integer
fifa.Height = fifa.Height.str.replace('cm','')
fifa.Height.unique()
```

```python
def convert_height(height):
    if height[-1] == '"':
        height = height.replace('\"', '')
        feet, inches = height.split("'")
        feet_to_cm = float(feet) * 30.48
        inches_to_cm = float(inches) * 2.54
        total_cm = feet_to_cm + inches_to_cm
        return int(total_cm)
    else:
        return int(height)
```

```python
fifa.Height = fifa.Height.apply(convert_height)
```

```python
fifa.Height.unique()
```

```python
#convert lbs to kg
def convert_lbs(w):
    if 'lbs' in w:
        w = w.replace('lbs','')
        w_to_kg = int(w) * 0.453
        return int(w)
    else:
        return int(w)
```

```python
fifa.Weight = fifa.Weight.apply(convert_lbs)
```

```python
#convert 'Value' column to integer
fifa.Value = fifa.Value.str.replace('€','')
fifa.Value.unique()
```

```python
def value_clean(v):
    if 'M' in v:
        v = v.replace('M','')
        v = float(v) * 1000000
        return int(v)
    elif 'K' in v:
        v = v.replace('K','')
        v = float(v) * 1000
        return int(v)
    else:
        return int(v)
```

```python
fifa.Value = fifa.Value.apply(value_clean)
```
